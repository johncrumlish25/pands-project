## Import Libraries  
In this section, I import the Python libraries used in this project.  
Pandas is used for handling and analysing the dataset, while Matplotlib is used to create visualisations such as histograms and scatter plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading the Dataset  
In this section, I load the Iris dataset from a CSV file into a pandas DataFrame.  
As the dataset does not include column names, they are assigned manually to improve clarity and readability.

```python
# Load the dataset from the CSV file
# header=None is used because the dataset does not include column names
df = pd.read_csv("iris.csv", header=None)

# Assign column names for better readability
df.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]

# Display the first 5 rows to check that the data loaded correctly
df.head()
```

## Summary of the Dataset  
In this section, I generate summary statistics for each numerical variable in the dataset using pandas.  
These statistics include measures such as the mean, minimum, maximum, and standard deviation.  
The results are also written to a text file (summary.txt) for later reference.

```python
# Generate summary statistics for the dataset
summary = df.describe()

# Display the summary in the notebook
print(summary)

# Write the summary to a text file
with open("summary.txt", "w") as f:
    f.write(str(summary))
```

## Histograms of Variables  
[Matplotlib Histograms](https://www.w3schools.com/python/matplotlib_histograms.asp)

In this section, I create histograms for each numerical variable in the dataset using Matplotlib.  
These histograms show how the values of each variable are distributed and help to identify patterns such as spread and variation.  
Each histogram is also saved as an image file for reference.

```python
# List of numerical columns (exclude species)
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

# Loop through each column and create a histogram
for col in columns:
    plt.figure()  # Create a new figure for each plot
    
    # Create histogram
    plt.hist(df[col])
    
    # Add title and labels
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    
    # Save the plot as a PNG file
    plt.savefig(f"{col}_hist.png")
    
    # Display the plot
    plt.show()
```

The histograms show how the values of each variable are distributed.  
For example, petal length has a wider spread than sepal width, indicating greater variation in petal measurements.  
This suggests that petal features may be more useful for distinguishing between species.

## Scatter Plots of Variables  
[Matplotlib Scatter Plots](https://www.w3schools.com/python/matplotlib_scatter.asp)

In this section, I create scatter plots to examine the relationship between each pair of variables.  
These plots help to identify patterns and differences between species.

```python
# List of numerical columns
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

# Loop through each pair of variables
for x_index in range(len(columns)):
    for y_index in range(x_index + 1, len(columns)):
        
        x = columns[x_index]
        y = columns[y_index]
        
        plt.figure()
        
        # Create scatter plot
        plt.scatter(df[x], df[y])
        
        # Add labels and title
        plt.xlabel(x)
        plt.ylabel(y)
        plt.title(f"{x} vs {y}")
        
        # Save the plot
        plt.savefig(f"{x}_vs_{y}.png")
        
        # Show the plot
        plt.show()
```

The scatter plots show relationships between variables.  
Petal length and petal width show a strong positive relationship, while sepal measurements are more spread out.

## Additional Analysis 

[groupby()](https://www.w3schools.com/python/pandas/ref_df_groupby.asp)

[corr()](https://www.w3schools.com/python/pandas/ref_df_corr.asp)

In this section, I perform further analysis to better understand the relationships between variables and differences between species.

```python
# Calculate the average values for each species
species_means = df.groupby("species").mean()

# Display the result
species_means
```

The table shows the average measurements for each species, highlighting setosa has noticeably smaller petal length and width compared to the other species

```python
# Calculate correlation between variables
correlation = df.drop("species", axis=1).corr()

# Display correlation matrix
correlation
```

The correlation matrix shows how strongly the numerical variables are related to each other.  
Values closer to 1 indicate a strong positive relationship, meaning both variables increase together, while values closer to 0 indicate little to no relationship.

From the results, petal length and petal width have a very strong positive correlation, suggesting that as the length of the petal increases, the width also increases.
