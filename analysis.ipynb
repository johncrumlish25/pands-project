## Import Libraries
In this section, I import the Python libraries needed for the analysis.
Pandas is used to analyse the dataset, while Matplotlib is used to create visualisations such as histograms and scatter plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading the Dataset  
In this section, I load the Iris dataset from a CSV file into a pandas DataFrame. 
Since the dataset does not include column names, I assign them manually for clarity.

```python
# Load the dataset from the CSV file
# header=None is used because the dataset does not include column names
df = pd.read_csv("iris.csv", header=None)

# Assign column names for better readability
df.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]

# Display the first 5 rows to check that the data loaded correctly
df.head()
```

## Summary of the Dataset  
In this section, I generate summary statistics for each variable in the dataset.  
These statistics are also written to a text file (summary.txt) for later reference.

```python
# Generate summary statistics for the dataset
summary = df.describe()

# Display the summary in the notebook
print(summary)

# Write the summary to a text file
with open("summary.txt", "w") as f:
    f.write(str(summary))
```
